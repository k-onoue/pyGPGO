```python
import torch
import plotly.graph_objects as go
import numpy as np

# SinusoidalSyntheticクラスのインスタンスを作成
sinusoidal_function = SinusoidalSynthetic()

# プロット用のxの範囲を生成
x_values = torch.linspace(5, 10, 500)  # 5から10まで500点を生成
y_values = sinusoidal_function(x_values) # yの値を計

x_values = x_values.squeeze().detach().numpy()
y_values = y_values.squeeze().detach().numpy()

# Plotlyでプロット
fig = go.Figure()

fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='lines', name='f(x) = sin(x)'))

# プロットのレイアウトを設定
fig.update_layout(
    title="Sinusoidal Synthetic Function",
    xaxis_title="x",
    yaxis_title="f(x)",
    template="plotly_white"
)

fig.show()
```
